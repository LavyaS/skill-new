```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import IsolationForest
from matplotlib.ticker import StrMethodFormatter
from IPython.display import display

# Load dataset
data = pd.read_csv('/housing.csv')

# Convert column names to lowercase
data.columns = data.columns.str.lower().str.strip()

# Display basic information
print(f"Dataset Size: {data.size}")
print(f"Dataset Shape: {data.shape}")
data.info()
display(data.head())

# Check for missing values
print("Missing Values (NAN):")
print(data.isna().sum()[data.isna().sum() > 0])

print("\nMissing Values (NULL):")
print(data.isnull().sum()[data.isnull().sum() > 0])

# Identify numerical and categorical columns
numeric_data = data.select_dtypes(exclude=['object']).columns.tolist()
categorical_data = data.select_dtypes(include=['object']).columns.tolist()

print("\nNumerical Features:", numeric_data)
print("Categorical Features:", categorical_data)

# Remove 'address' column (unique for each row, doesn't contribute to modeling)
if 'address' in categorical_data:
    data = data.drop(columns=['address'])
    categorical_data.remove('address')

# Convert categorical values to lowercase and strip spaces
data[categorical_data] = data[categorical_data].apply(lambda x: x.str.lower().str.strip())

# Function to remove outliers using IQR
def remove_outliers(data):
    df_clean = data.copy()
    for col in numeric_data:
        Q1 = df_clean[col].quantile(0.15)
        Q3 = df_clean[col].quantile(0.85)
        IQR = Q3 - Q1
        df_clean = df_clean[(df_clean[col] >= Q1 - 1.5 * IQR) & (df_clean[col] <= Q3 + 1.5 * IQR)]
    return df_clean

data = remove_outliers(data)

# Function to remove outliers using Isolation Forest
def isolation_forest_outliers(data):
    df_clean = data.copy()
    for col in numeric_data:
        column_data = df_clean[col].values.reshape(-1, 1)
        clf = IsolationForest(n_estimators=200, max_samples=0.7, contamination=0.005, random_state=1)
        preds = clf.fit_predict(column_data)
        df_clean = df_clean[preds != -1]
    return df_clean

data = isolation_forest_outliers(data)

# Encode categorical variables
encoder = LabelEncoder()
for col in categorical_data:
    data[col] = encoder.fit_transform(data[col])

# Standardize numerical features
def scale_data(data):
    scaler = StandardScaler()
    data[numeric_data] = scaler.fit_transform(data[numeric_data])
    return data

data = scale_data(data)

# Correlation Heatmap
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()

# Feature Selection based on correlation threshold
def correlation(x, y, threshold=0.7):
    return [col for col in x.columns if abs(np.corrcoef(x[col], y)[0, 1]) > threshold]

if 'price' in data.columns:
    x = data.drop(columns=['price'])
    y = data['price']
    to_drop = correlation(x, y, 0.7)
    data = data.drop(columns=to_drop, errors='ignore')
    print("Dropped Features Due to High Correlation:", to_drop)

# --- Regression Models ---
# Simple Linear Regression
if 'price' in data.columns and 'area population' in data.columns:
    x_simple = data[['price']]
    y_simple = data[['area population']]
    x_train, x_test, y_train, y_test = train_test_split(x_simple, y_simple, test_size=0.2, random_state=1)

    model_simple = LinearRegression()
    model_simple.fit(x_train, y_train)
    score_simple = model_simple.score(x_test, y_test)
    print(f"Simple Linear Regression R Score: {score_simple:.2f}")

    plt.figure(figsize=(8,5))
    plt.scatter(x_simple, y_simple, color='black', marker='.')
    plt.plot(x_simple, model_simple.predict(x_simple), color='blue')
    plt.xlabel('Price')
    plt.ylabel('Area Population')
    plt.title('Simple Linear Regression')
    plt.gca().xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
    plt.show()

    rss_simple = np.mean((model_simple.predict(x_simple) - y_simple) ** 2)
    print(f"Residual Sum of Squares (RSS): {rss_simple:.2f}")
else:
    print("Error: 'price' or 'area population' column is missing. Check preprocessing.")

# Multiple Linear Regression
if 'price' in data.columns:
    x_multi = data.drop(columns=['price'])
    y_multi = data['price']
    x_train, x_test, y_train, y_test = train_test_split(x_multi, y_multi, test_size=0.2, random_state=1)

    model_multi = LinearRegression()
    model_multi.fit(x_train, y_train)
    score_multi = model_multi.score(x_test, y_test)
    print(f"Multiple Linear Regression R Score: {score_multi:.2f}")

    rss_multi = np.mean((model_multi.predict(x_multi) - y_multi) ** 2)
    print(f"Residual Sum of Squares (RSS): {rss_multi:.2f}")
else:
    print("Error: 'price' column is missing. Check preprocessing.")



```
